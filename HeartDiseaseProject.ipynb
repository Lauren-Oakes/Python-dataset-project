# Heart Disease Data Analysis Project

#### This dataset is taken from the Framingham Heart Study that was designed to identify risk factors for cardiovascular disease. It began in 1948 with 5209 participants and is still ongoing today. This subset of data is clinic data from 4434 participants over 3 examination periods 6 years apart (1956 - 1968). Participants were followed for a total of 24 years for outcomes.

## 3. Display the data

#### Display data using a combination of count plots and histograms.

```python
slimmed_df = df.drop (['PREVCHD', 'PREVAP', 'PREVMI', 'PREVSTRK', 'PREVHYP', 'TIME'], axis = 1)
var_cat = ['SEX','CURSMOKE','DIABETES','BPMEDS','educ','PERIOD','DEATH','ANGINA','HOSPMI',
           'MI_FCHD','ANYCHD','STROKE','CVD','HYPERTEN']

fig, ax = plt.subplots(11, 3, figsize=(16, 32))

a = 0
b = 0
for var in slimmed_df.columns:
    if var in var_cat:
        sns.countplot(x = var, data = slimmed_df, color = 'teal', ax=ax[a,b])
    else:
        sns.histplot(slimmed_df[var], color = 'teal', edgecolor = 'black', ax=ax[a,b])
    ax[a,b].set_title(var)
    ax[a,b].set_xlabel(var)
    ax[a,b].set_ylabel('Count')

    if b==2:
        b=0
        a+=1
    else:
        b+=1

plt.tight_layout()
plt.show()
```

#### Find out how many participants died in the 24 year follow up period.

0 indicates the participant was still living and 1 indicates they had died. Data also filtered to only show Period 1 data to avoid duplicates.

```python
df_period1 = df[df['PERIOD'] == 1]
df_period1['DEATH'].value_counts()
```

#### Find out how many participants developed CHD

```python
df_period1['ANYCHD'].value_counts()
```

#### Comparison of deaths and CHD

```python
death_count_period1 = df_period1[df_period1['DEATH'] == 1].shape[0] 
chd_count_period1 = df_period1[df_period1['ANYCHD'] == 1].shape[0]  

summary_period1 = pd.DataFrame({
    'Deaths': [death_count_period1],
    'Participants with CHD': [chd_count_period1]
})

summary_period1.plot(kind='bar', figsize=(6, 4), color=['salmon', 'skyblue'])

plt.title('Number of Deaths vs. Participants with CHD')
plt.xlabel('Metrics')
plt.ylabel('Count')
plt.xticks(rotation = 0)
plt.tight_layout()

plt.show()
```

## 4. Correlation Analysis

#### Performed a Pearson correlation plot using some of the attributes.

0.4 -0.7 is a moderate correlation and above 0.7 is a strong correlation.

```python
corr_df = df.drop(['RANDID', 'PREVCHD', 'PREVAP', 'PREVMI', 'PREVSTRK', 'PREVHYP','TIMEAP', 'TIMEMI', 'TIMEMIFC', 'TIMECHD',
       'TIMESTRK', 'TIMECVD', 'TIMEDTH', 'TIMEHYP', 'PERIOD', 'TIME'], axis=1)
corr_df.head()

cor = corr_df.corr(method='pearson')
fig, ax = plt.subplots(figsize=(8, 6))
plt.title("Correlation Plot")
sns.heatmap(cor)
plt.show()
```

## 5. Data Analysis

#### Plot HDL/LDL against Total Cholesterol to show relationships

```python
df.plot.scatter(x = 'TOTCHOL', y = 'HDLC',s=1, c='teal')
df.plot.scatter(x = 'TOTCHOL', y = 'LDLC',s=1, c='teal')

plt.tight_layout()
```

#### Pairplots for HDL/LDL in participants with CHD

```python
sns.pairplot(df[['LDLC', 'HDLC','ANYCHD']], hue='ANYCHD')
plt.show()
```

#### Boxplots for several attributes against whether the participant developed CHD

```python
fig, axes = plt.subplots(2, 3, figsize=(14,10))

sns.boxplot(x='ANYCHD', y='TOTCHOL', data=df, ax=axes[0,0])
axes[0, 0].set_title('Total Cholesterol vs. CHD Outcome')

sns.boxplot(x='ANYCHD', y='BMI', data=df, ax=axes[0,1])
axes[0, 1].set_title('BMI vs. CHD Outcome')

sns.boxplot(x='ANYCHD', y='SYSBP', data=df, ax=axes[0,2])
axes[0,2].set_title('Systolic BP vs. CHD Outcome')

sns.boxplot(x='ANYCHD', y='DIABP', data=df, ax=axes[1,0])
axes[1,0].set_title('Diastolic BP vs. CHD Outcome')

sns.boxplot(x='ANYCHD', y='HDLC', data=df, ax=axes[1,1])
axes[1, 1].set_title('HDL Cholesterol vs. CHD Outcome')

sns.boxplot(x='ANYCHD', y='LDLC', data=df, ax=axes[1,2])
axes[1, 2].set_title('LDL Cholesterol vs. CHD Outcome')

plt.tight_layout()
plt.show()
```
